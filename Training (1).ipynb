```python
import pandas as pd
import numpy as np
import matplotlib
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from collections import Counter
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import f1_score
```

```python
df=pd.read_csv('train.csv')
```

```python
label_encoder = preprocessing.LabelEncoder()
df['work_type']= label_encoder.fit_transform(df['work_type'])
  
```

```python

df['gender']= label_encoder.fit_transform(df['gender'])
df['ever_married']= label_encoder.fit_transform(df['ever_married'])
df['Residence_type']= label_encoder.fit_transform(df['Residence_type'])
df['smoking_status']= label_encoder.fit_transform(df['smoking_status'])

df['blood'].fillna('AB',inplace=True)
df['blood']= label_encoder.fit_transform(df['blood'])
```

```python
df.info()
```

```python
df['age'].fillna(value=43.071256,inplace=True)
df['heart_disease'].fillna(0,inplace=True)
df['avg_glucose_level'].fillna(106.178062,inplace=True)
df['bmi'].fillna(28.887899,inplace=True)
df['height'].fillna(1.611831,inplace=True)
df['hypertension'].fillna(0,inplace=True)
```

```python
df.isnull().sum()  #check null
```

```python
df.info()
```

```python
x = df[['gender','age','hypertension','heart_disease','ever_married','work_type','Residence_type',
        'avg_glucose_level','bmi','smoking_status','blood','height']]
y= df['stroke']

counter = Counter(y)
print(counter)

#Phân loại không cân bằng với Python
over = SMOTE()
x, y = over.fit_resample(x, y)

#under = RandomUnderSampler(sampling_strategy=0.5)
#x, y = under.fit_resample(x, y)

counter = Counter(y)
print(counter)

x_train, x_test, y_train, y_test= train_test_split(x,y,random_state=0)

lg= LogisticRegression()
lg.fit(x_train,y_train)
test=lg.score(x_test,y_test)
print(test)

pre=lg.predict(x_test)
print(pre)

f1_score(y_test,pre)
```
